## Cleaning NoseTouch Data for rendering in d3



```python
import pandas as pd
import re
from datetime import datetime
```

```python
nosetouch_data = pd.read_csv("nosetouch_data.csv")
nosetouch_data.head()
```

Done
- remove note column
- add underscores to column names
- convert time and date to datetime format 

```python
#remove note column
nosetouch_data = nosetouch_data.drop(columns=["Note"])
#all column headers to lowercase, with underscores instead of spaces, remove colons
nosetouch_data.columns = [re.sub(r'[:\s]+', '_', col.lower()) for col in nosetouch_data.columns]
nosetouch_data.head()
```

```python
type(nosetouch_data["date"][0]) #string format, so I need to use datetime to convert it to a date format

#create new datetime column from Date and Time columns
    #format for "Date" us MM DD, format for "Time" is HH:MM, year is 2026
nosetouch_data["datetime"] = nosetouch_data.apply(
    lambda row: datetime.strptime(f"{row['date'].strip()} 2026 {row['time'].strip()}", "%b %d %Y %H:%M"), 
    axis=1
)
nosetouch_data.head()

```

```python
#export prepared data to new csv file
nosetouch_data.to_csv("nosetouch_data_prepared.csv", index=False)
```
